```python
import matplotlib.pyplot as plt
import matplotlib.transforms as transf
```

```python
class Rect(object):
    
    def __init__(self,cx,cy,width,height,color,transp,lw):
        self.cx = cx
        self.cy = cy
        self.width = width
        self.height = height
        self.color = color
        self.transp = transp
        self.lw = lw
        
        w2 = self.width/2.0
        h2 = self.height/2.0
        #list of x and list of y coordinates
        self.x = [self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2,self.cx-w2]
        self.y = [self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2,self.cy-h2]
        
        
    def display(self):
        plt.axis('equal')
        plt.plot(self.x,self.y,c=self.color,linewidth=self.lw,alpha=self.transp)
        plt.fill()
        #plt.fill(self.x,self.y,self.color,c="k",linewidth)

    def display2(self):
        plt.axis('equal')
        plt.fill(self.x,self.y,c=self.color,linewidth=self.lw,alpha=self.transp)
        
        
```

```python
# rectangle center pt, width, height

r = Rect(1,2,3,4,[0,0,0],1,1)
r.display()
```
