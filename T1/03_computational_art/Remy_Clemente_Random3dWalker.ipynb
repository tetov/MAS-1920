```python
import matplotlib.pyplot as plt
import math
import random

class Line:
    def __init__ (self, start_x,start_y,iteration, width, height):
        self.it = iteration
        self.w = width
        self.h = height
        self.list_x = []
        self.list_y = []
        self.list_x.append(start_x)
        self.list_y.append(start_y)
        self.display_x = []
        self.display_y = []
        
    def generate(self):
        
        #Random points in space
        for i in range(self.it):
            xp = random.uniform(0,self.w)
            yp = random.uniform(0,self.h)
            self.list_x.append(xp)
            self.list_y.append(yp)
            
        #Random paths from i to i+1
        paths = []
        
        for i in range(self.it):
            generator = random.random()
            
            if generator <= 0.33:
                path = 0
                self.display_x.append(self.list_x[i])
                self.display_y.append(self.list_y[i])
                self.display_x.append(self.list_x[i+1])
                self.display_y.append(self.list_y[i])

            elif generator <= 0.66:
                path = 1
                self.display_x.append(self.list_x[i])
                self.display_y.append(self.list_y[i])
                self.display_x.append(self.list_x[i])
                self.display_y.append(self.list_y[i+1])                
                
            else:
                path = 2
                factor = random.random()/2
                self.display_x.append(self.list_x[i])
                self.display_y.append(self.list_y[i])
                self.list_x[i+1]=(((self.w/2)-(self.list_x[i]))*factor)+self.list_x[i]
                self.list_y[i+1] =(((self.h/2)-(self.list_y[i]))*factor)+self.list_y[i]
                self.display_x.append(self.list_x[i+1])
                self.display_y.append(self.list_y[i+1])
                
            paths.append(path)
        print(paths)

    def display(self):
        #display figure
        plt.figure(figsize=(10,10))
        plt.plot(self.display_x, self.display_y,c=[0,0,0],linewidth=0.5)

```

```python
width = 100
height = 100
iteration_num = 100


x = random.random()*width
y = random.random()*height
        
l = Line(x,y,iteration_num,width, height)
l.generate()
l.display()
```
