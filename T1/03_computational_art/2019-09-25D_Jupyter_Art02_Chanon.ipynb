```python
import matplotlib.pyplot as plt
```

```python
class Rect(object):
    def __init__(self, cx, cy, width, height):
        self.cx=cx
        self.cy=cy
        self.width=width
        self.height=height
        w2=self.width/2.0
        h2=self.height/2.0
        # list of x and list of y coordinates
        self.x=[self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y=[self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
                
            
    def display(self):
        plt.axis("equal")
        plt.plot(self.x, self.y, c="k", linewidth=1)
        #plt.fill(self.x, self.y, c=cv, linewidth=1)
```

```python
plt.figure(figsize=(12,12))
rects=[]
for x in range(5):
    for y in range(5):
        r=Rect(x*10, y*10, 8, 8)
        rects.append(r)
for r in rects:
    r.display()
```

```python
import math

class Rect(object):
    def __init__(self, cx, cy, width, height):
        self.cx=cx
        self.cy=cy
        self.width=width
        self.height=height
        w2=self.width/2.0
        h2=self.height/2.0
        # list of x and list of y coordinates
        self.x=[self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y=[self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
                
    def rotate(self, rot):
        self.new_x = []
        self.new_y = []
        for i, px in enumerate(self.x):
            qx = self.cx + math.cos(rot) * (px - self.cx) - math.sin(rot) * (self.y[i]  - self.cy)
            self.new_x.append(qx)
        for i, py in enumerate(self.y):
            qy = self.cy + math.sin(rot) * (self.x[i] - self.cx) + math.cos(rot) * (py - self.cy)
            self.new_y.append(qy)
    
    def display(self):
        plt.axis("equal")
        plt.plot(self.new_x, self.new_y, c="k", linewidth=1)
```

```python
plt.figure(figsize=(12,12))
rects=[]
for x in range(5):
    for y in range(5):
        for i in range(8):
            s=i+1
            ox=2*random.random()-2
            oy=5*random.random()-2
            r=Rect(x*10+ox, y*10+oy, s, s)
            
            #r=Rect(x*10, y*10, s*2, s*2)
            rects.append(r)
            r.rotate(y/2)
for r in rects:
    r.display()

```
