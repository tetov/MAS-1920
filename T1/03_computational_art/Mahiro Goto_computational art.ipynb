```python
class Rect(object):
    def __init__(self, cx, cy, width, height):
        self.cx = cx
        self.cy = cy
        self.width = width
        self.height = height
        w2 = self.width/2.0
        h2 = self.height/2.0
        # list of x and list of y coordinates
        self.x = [self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y = [self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
        
    def display(self, cv):
        plt.axis('off')
        # plt.plot(self.x, self.y, c='k', linewidth=1)
        plt.plot(self.x, self.y, c=cv, linewidth=3+random.random())
        plt.fill(self.x, self.y, c=cv, linewidth=1)
        
```

```python
plt.figure(figsize=(12,12))
plt.axis('off')
rects = []
for x in range(5):
    for y in range(5):
        for i in range(5):
            s = 3+random.random()*6
            ox = 3*random.random()-1.5
            oy = 3*random.random()-1.5
            # print(oy, ox)
            r = Rect(x*10 + ox, y*10 + oy, s, s)
            rects.append(r)

for r in rects:
    ct = [random.random(), 0.2, 0, 0.6]
    r.display(ct)
```
