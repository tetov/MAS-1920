```python
class Rect(object):
    def __init__(self, cx, cy, width, height):
        self.cx=cx
        self.cy=cy
        self.width=width
        self.height=height
        w2= self.width/2.0
        h2= self.height/2.0
        # list of x and list of y coordinates
        self.x= [self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y= [self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
        
    def display(self, cv):
        plt.axis('equal')
        #plt.plot(self.x, self.y, c="k", linewidth=1)
        plt.fill(self.x, self.y, c=cv, linewidth= random.random()*2)
```

```python
# rectangle centre pt, width, height
r= Rect(1,2,3,4)
r.display()
```

```python
import matplotlib.pyplot as plt
import random
import math


class Rect(object):
    def __init__(self, cx, cy, width, height):
        self.cx=cx
        self.cy=cy
        self.width=width
        self.height=height
        w2= self.width/2.0
        h2= self.height/2.0
        # list of x and list of y coordinates
        self.x= [self.cx-w2, self.cx, self.cx, self.cx+w2, self.cx+w2]
        self.y= [self.cy-h2, self.cy-h2, self.cy, self.cy, self.cy+h2]
        
    def display(self, cv):
        plt.axis('equal')
        #plt.plot(self.x, self.y, c="k", linewidth=1)
        plt.plot(self.x, self.y, c=cv, linewidth= random.random()*2)


plt.figure(figsize=(12,12))
plt.axis("off")
rects= []
for x in range(10):
    for y in range(10):
        for i in range(10):
            s= 3+random.random()*6
            ox= 50*random.random() +2 #to offset the centre
            oy= 50*random.random() +2
            #print (oy, ox)
            r=Rect(x*20 + ox, y*20 + oy, s, s) 
            rects.append(r) 
            
    
for r in rects:
    ct = [random.random(),0.8,0.9]
    r.display(ct)
```
