```python
import matplotlib.pyplot as plt
```

```python
class Rect(object):
    def __init__(self, cx, cy, width, height):
        
        self.cx = cx
        self.cy = cy
        self.width = width
        self.height = height
        w2 = self.width/2.0
        h2 = self.height/2.0
        #list of x and list of y coordinates
        self.x = [self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y = [self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
        
        
    def display(self,cv):
        plt.axis("equal")
        #plt.plot(self.x, self.y, c="k", linewidth=1)
        plt.fill(self.x, self.y, c=cv, linewidth=0)
```

```python
#rectangle center pt, width ,height
r = Rect(1,2,3,4)
r.display()
```

```python
import math
class Rect(object):
    def __init__(self, cx, cy, width, height, res, ang):
        
        self.cx = cx
        self.cy = cy
        self.width = width
        self.height = height
        self.ang = ang
        w2 = self.width/2.0
        h2 = self.height/2.0
        a=2*math.pi/res
        #list of x and list of y coordinates
        
        self.x=[]
        self.y=[]
        
        for i in range(res):
            self.x.append(cx+math.cos(i*a+ang)*w2)
            self.y.append(cy+math.sin(i*a+ang)*h2)
            
        
        
        #print(self.x)
        
#         self.x = [cx+math.cos(0*a+ang)*w2, cx+math.cos(1*a+ang)*w2, cx+math.cos(2*a+ang)*w2, cx+math.cos(3*a+ang)*w2, cx+math.cos(0*a+ang)*w2]
#         self.y = [cy+math.sin(0*a+ang)*h2, cy+math.sin(1*a+ang)*h2, cy+math.sin(2*a+ang)*h2, cy+math.sin(3*a+ang)*h2, cy+math.sin(0*a+ang)*h2]
        
        
    def display(self,cv):
        plt.axis("equal")
        #plt.plot(self.x, self.y, c="k", linewidth=1)
        plt.fill(self.x, self.y, c=cv, linewidth=0)
        
    def display_l(self):
        plt.axis("equal")
        plt.plot(self.x, self.y, c="k", linewidth=0.5)
```
