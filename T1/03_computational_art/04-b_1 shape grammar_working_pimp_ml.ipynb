```python
class algae(object):
    def __init__(self,ruleA=['A','B'],ruleB=['A'],gen=3,r=0,x=0,y=0,nx=0,ny=1,Ax=0,Ay=.6,Bx=.6,By=.4,c='k',lw=1):
        self.gen=gen  #genlength
        self.RA=ruleA
        self.RB=ruleB
        self.x=x
        self.y=y
        self.nx=nx
        self.ny=ny
        self.startW=lw  #startlinewidth
        self.r=r       #modulo of random value to add
        self.Ax=Ax    #transformation of A branch
        self.Ay=Ay
        self.Bx=Bx    #transformation of B branch
        self.By=By
        self.c=c


        #shape_grammar_A--> ? B-->??  ##########################################################
        
        print('''
        #rules:
        # 'A' -> ''',self.RA,'''
        # 'B' -> ''',self.RB,'''
        ''')
        #axiom: 'B'
        gen_list = ['A']

        full_gen=gen_list


        #print('axiom:  ',gen_list)
        for i in range(self.gen):

            X = len(gen_list)
            next_gen = []
            for l in range(X):
                c = gen_list[l] #c(current item)

                if c == 'B':
                    n = self.RB
                else:
                    n = self.RA
                next_gen = next_gen+n

            gen_list = next_gen
            full_gen.append(gen_list)
     

        #execute the rules ###################################

        #current positions list
        px = [self.x,self.x+self.nx]
        py = [self.y,self.y+self.ny]
        #start line
        line =[self.x, self.y, self.x+self.nx, self.y+self.ny]
        #next generation positions list
        gx=[]
        gy=[]
        
        for g in range(len(full_gen)):
            current_gen=full_gen[g]
            #branch behavior, random or not
            if self.r==0:
                M=(g+1)
            else:
                M=(g+1)*(random.random()-.5)*self.r
                
            for r in range(len(current_gen)):
                
                rule = current_gen[r]
                
                # rule A action
                if rule == 'A':
                    x = px.pop(-1)
                    y = py.pop(-1)
                    nx= x + self.Ax/M
                    ny= y + self.Ay/M


                    nline =[x,y,nx,ny]
                    line.extend(nline)

                    gx.insert(0,nx)
                    gy.insert(0,ny)
                    gx.insert(0,nx)
                    gy.insert(0,ny)

                #rule B action
                elif rule == 'B':
                    x = px.pop(-1)
                    y = py.pop(-1)
                    nx= x + self.Bx/M
                    ny= y + self.By/M

                    nline =[x,y,nx,ny]
                    line.extend(nline)

                    gx.insert(0,nx)
                    gy.insert(0,ny)
        
            #overwrite the position list for the next gen
            px=gx
            py=gy

        #return line coordinates
        pointList=line
        

        #draw the algae #########################################

        plt.axis('equal')

        n = 4
        a = int(len(pointList)/n)
        for i in range(a):
            sub = pointList[i*n:(i*n)+n]


            x = sub[0::2]
            y = sub[1::2]
            c=self.c
            plt.plot(x,y,c,linewidth=self.startW)
```
