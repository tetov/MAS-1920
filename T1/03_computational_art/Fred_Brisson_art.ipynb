```python
import matplotlib.pyplot as plt
import math
import random
# this is to import math and random library from the default

class Circle:
    def __init__(self, rad, xpos, ypos, color, rot): #this is to define the self value of each of the circle
        # store init values as internal variables
        # all default argument (origin = (0,0,0)) like the one above has to be at the end
        self.rad = rad
        self.xpos = xpos
        self.ypos = ypos
        self.color = color
        self.rot=rot
        
    def display(self): #this function is to plot the circle
        xc=[]
        yc=[]
        res = 4
        radian=self.rot*math.pi/180
        a=2*math.pi/res
        for i in range(res):
            xc.append(self.xpos+math.cos(i*a-(a/2)+radian)*self.rad)
            yc.append(self.ypos+math.sin(i*a-(a/2)+radian)*self.rad)
        plt.fill(xc,yc,c=self.color)
        # the line above is to plot the circle.

plt.figure(figsize=(12,12)) # this comes from the matplotlib
rects=[] # this is an empty list, this is where we draw based (with chosen size, colour, rotation, location in this block)
for x in range(5): # those are the columns
    for y in range(5): # those are the rows
        s=1
        r=Circle(s*5,x*10,y*10,[1,0,0],random.randint(0,90))
        rects.append(r) # here we add the value r which represents the Circle
            
for r in rects: #here we put the value r (as a list) back into the function rects
    r.display() #here we display the value r
```
