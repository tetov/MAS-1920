```python
import matplotlib.pyplot as plt
import math
import random
```

```python
class Triangle: 
    def __init__(self, ox, oy, base, height, alpha):
        self.ox = ox
        self.oy = oy
        self.b = base
        self.h = height
        self.a = alpha
     
        
        #list of x and y coordinates
        self.x = [self.ox, self.ox + self.b, self.ox + self.h/math.tan(math.pi*self.a/180), self.ox ]
        self.y = [self.oy, self.oy, self.oy + self.h, self.oy]
        
    def display(self):
        plt.axis('equal')
        plt.plot(self.x, self.y, c='k', linewidth=1)
```

```python
class Triangle: 
    def __init__(self, ox, oy, base, height, alpha, beta):
        self.ox = ox
        self.oy = oy
        self.b = base
        self.h = height
        self.a = alpha
        self.be = beta
     
        
        #list of x and y coordinates
        self.x = [self.ox, self.ox + self.b*math.cos(math.pi*self.be/180), self.ox + self.h/math.sin(math.pi*self.a/180)*math.cos(self.a + self.be), self.ox]
        self.y = [self.oy, self.oy + self.b*math.sin(math.pi*self.be/180), self.oy + self.h/math.sin(math.pi*self.a/180)*math.sin(self.a + self.be), self.oy]
        
    def display(self,cv,cf):
        plt.axis('equal')
        #plt.plot(self.x, self.y, c='k', linewidth=1)
        plt.plot(self.x, self.y, c=cv, linewidth=2)
        plt.fill(self.x, self.y, c=cf)
```

```python
class Circle:
    #def __init__(self,radius,origin = (0,0,0)): #if i don't specify the origin, origin will be 0,0,0
    def __init__(self,radius,xpos, ypos, color): #here the origin is defined with xpos and ypos
        #store init values as internal variables (self.radius for ex.)
       
        self.radius = radius
        self.xpos = xpos
        self.ypos = ypos
        self.color = color
        #make 2 functions

    def calculate_circumference(self):
        perimeter=math.pi*2*self.radius
        print("circumference is", perimeter)
        return perimeter

    def display(self):
        xc=[]
        yc=[]
      
        res=4 #resolution of the circle
        a=2*math.pi/res
        for i in range (res):
            xc.append(self.xpos + math.cos(i*a-a/2)*self.radius)
            yc.append(self.ypos + math.sin(i*a-a/2)*self.radius)

            
        plt.fill(xc,yc,c=self.color)
        
      
c1=Circle(5,0.5,0.5,[1,0,1]) 
plt.figure(figsize=(6,6))
plt.axis('equal')
c1.display()

#pts_x= c1.xc
#pts_y= c1.yc
```
