```python
import matplotlib.pyplot as plt
import math
```

```python
class Rect:
    def __init__(self, cx,cy,width,height):
        self.cx = cx
        self.cy = cy
        self.width = width
        self.height = height
        
        w2 = self.width/2.0
        h2 = self.height/2.0
        #list of x and list of y coordinates
        self.x = [self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y = [self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
        
    def display (self):
        plt.axis("equal")
        plt.plot(self.x, self.y, c="k", linewidth=1)
```

```python
# rectangle center pt, width, height
r=Rect(1,2,3,4)
r.display()
```

```python
rects=[]
for x in range(5):
   
    for y in range(10):
        shift=x*y
        r= Rect (x*10,y*10+shift,8,8)
        rects.append(r)
        
for r in rects:
    r.display()
```
