```python
import matplotlib.pyplot as plt
import math
import random as r
```

```python
#plotter
plt.axis('equal')

def draw_line(pointList):
    
    n = 4
    a = int(len(pointList)/n)
    for i in range(a):
        sub = pointList[i*n:(i*n)+n]
     

        x = sub[0::2]
        y = sub[1::2]
    
        plt.plot(x,y)
```
