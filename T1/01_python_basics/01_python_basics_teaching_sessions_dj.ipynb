```python
def list_sum(number_list):
    output = 0
    
    for number in number_list:
        output += number
    
    return output

def say_hello(name):
    print("goodmorning "+name)

for i in range(1):
    if i%2 == 0:
        say_hello("davud")
    else:
        say_hello("mas")
```

```python
def circle_area(radius):
    pi = 3.14
    
    area = pi * radius ** 2
    
    return area
```

```python
for i in range(10):
    print("i = ",i, "area = ", circle_area(i))

fib = [1,2,3,5,8,13]

for number in fib:
    print("number = ",number, "area = ", circle_area(number))   
```

```python
import math
import matplotlib.pyplot as plt

class Circle:
    def __init__(self, radius, x_pos, y_pos, color):
        # store init values as internal variables
        self.radius = radius
        self.x = x_pos
        self.y = y_pos
        self.c = color
    def calculate_area(self):
        # calculate and print out area
        area = math.pi * self.radius ** 2
        return area
    def calculate_circumference(self):
        # calculate and print out circumference
        circumference = 2 * math.pi * self.radius
        return circumference
    def display(self):
        xc = []
        yc = []
        res = 100
        a = 2*math.pi/res
        for i in range(res):
            xc.append(self.x + math.cos(i*a) * self.radius)
            yc.append(self.y + math.sin(i*a) * self.radius)
        plt.fill(xc,yc,c=self.c)
```

```python
import random as r

plt.figure(figsize = (15,15))
plt.axis('equal')

for i in range(100):
    c = Circle(r.random(), r.random()*10, r.random()*10, (0,0,r.random()))
    c.display()
```

```python
waves = []

for i in range(7):
    wave = [m.sin((i+1)*j*m.pi/50)/(i+1) for j in range(100)]
    waves.append(wave)

print(waves)
```

```python
print("type of waves: ", type(waves))
print("length of waves", len(waves))
print("type of waves[0][0]", type(waves[0][0]))
print("type of waves[0]", len(waves[-1]))
```
