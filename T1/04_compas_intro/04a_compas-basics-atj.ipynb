```python
import compas
from compas_plotters import MeshPlotter
from compas.datastructures import Mesh
from pathlib import Path, PureWindowsPath


a_mesh = Mesh()

print(a_mesh.attributes)
print(a_mesh.data)

# Define mesh

a = a_mesh.add_vertex() # no args -> 0,0,0
b = a_mesh.add_vertex(x=1.)
c = a_mesh.add_vertex(x=1., y=1.)
d = a_mesh.add_vertex(y=1.)

# Print info

f = a_mesh.add_face([a, b, c, d])

a_mesh.summary()

print("Vertices coordinates")
for xyz in a_mesh.get_vertices_attributes('xyz'):
    print(xyz)
    
print("\nEdges connections between vertices")    
for key in a_mesh.edges():
    print(key)

# Plot mesh
    
plotter = MeshPlotter(a_mesh)

plotter.draw_edges()
plotter.draw_vertices()
plotter.draw_faces()

plotter.show()

```

```python
#delauney_obj = PureWindowsPath(Path('resources/delaunay.obj'))

mesh3 = Mesh.from_obj(str(Path('./resources/delaunay.obj')))



plotter = MeshPlotter(mesh3)

plotter.draw_edges()
plotter.draw_vertices(text={key: str(key) for key in mesh3.vertices()})
plotter.draw_faces(text={key: str(key) for key in mesh3.faces()})

plotter.show()
```

```python
for key in mesh3.vertices():
    print(key, mesh3.vertex_neighbors(key, ordered=True))
    print(key, mesh3.vertex_faces(key, ordered=True))
```

```python
for key in mesh3.vertices():
    v_degree = mesh3.vertex_degree(key)
    if v_degree >= 7:
        print(key, v_degree)
```

```python
plotter = MeshPlotter(mesh3)

plotter.draw_vertices(
    facecolor={key: '#333333' for key in mesh3.vertices_on_boundary()},
    radius={key: mesh3.vertex_degree(key)*.1 for key in mesh3.vertices()})

plotter.draw_edges(
    color={key: '#cccccc' for key in mesh3.edges()},
    width={key: 1 for key in mesh3.vertices()})

plotter.draw_faces(
    facecolor={key: '#999999' for key in mesh3.faces()})

plotter.show()
```

```python
mesh4 = compas.datastructures.mesh_dual(mesh3, cls=None)

plotter = MeshPlotter(mesh4)

plotter.draw_vertices(
    facecolor={key: '#bbbbbb' for key in mesh4.vertices_on_boundary()},
    radius={key: mesh4.vertex_degree(key)*.1 for key in mesh4.vertices()})

plotter.draw_edges(
    color={key: '#cccccc' for key in mesh4.edges()},
    width={key: 1 for key in mesh4.vertices()})

plotter.draw_faces(
    facecolor={key: '#999999' for key in mesh4.faces()})

plotter.show()
```
